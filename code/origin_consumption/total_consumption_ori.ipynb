```python
import pandas as pd 
import numpy as np
```

```python
#read the eplusout.csv file
df = pd.read_csv('eplusout.csv')
df
```

```python
#select the columns of DistrictHeating and DistrictCooling
df1 = df[['DistrictHeating:Facility [J](Hourly)','DistrictCooling:Facility [J](Hourly)']]
df1
```

```python
#select the columns of Electricity
df2 = df[['Electricity:Facility [J](Hourly)']]
df2
```

```python
#concat df1 and df2
df3 = pd.concat([df1,df2],axis=1)
df3
```

```python
#remove the rows with NaN   
df4 = df3.dropna()
```

```python
df4
```

```python
#convert the unit from J to kWh
df4['DistrictHeating:Facility [J](Hourly)'] = df4['DistrictHeating:Facility [J](Hourly)']/3.6e6
df4['DistrictCooling:Facility [J](Hourly)'] = df4['DistrictCooling:Facility [J](Hourly)']/3.6e6
df4['Electricity:Facility [J](Hourly)'] = df4['Electricity:Facility [J](Hourly)']/3.6e6
df4
```

```python
#sum the columns
df5 = df4.sum(axis=0)
```

```python
df5
```

```python
#total consumption
total_consumption = df5.sum(axis=0)
total_consumption
```
